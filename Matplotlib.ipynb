```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x  ** 2
```

```python
 x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'r')

plt.subplot(1, 2, 2)
plt.plot(y,x,'b')
```

```python
#OOP
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([1,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Large Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
axes
```

## Figure Size and DPI

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('output/my_picture.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.plot(x*x,y, label='Squared')
ax.plot(x,y, label='Original')

ax.legend(loc=0)
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange', lw=1, alpha=0.5, ls='--', marker='o', ms=8, mfc='blue', mew=1)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red', lw=1, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
