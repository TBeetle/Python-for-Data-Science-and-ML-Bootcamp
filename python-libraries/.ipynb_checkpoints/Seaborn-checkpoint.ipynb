```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

## Categorial Plots

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tc =tips.corr(numeric_only=True)
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers', observed=False)
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white',linewidth=2)
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', hue='sex')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)
```

```python
sns.set_context('notebook',font_scale=1)
sns.countplot(x='sex', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='bone')
```
